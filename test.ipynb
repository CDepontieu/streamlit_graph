```python
from pyvis.network import Network
import networkx as nx

group_color_mapping = {
    0: 'red',
    1: 'blue',
    2: 'green'
}

# Initialize the Network
g = Network(select_menu=True, filter_menu=True, layout=True)#options={"layout": {"hierarchical": {"direction": "UD"}}})

# Add nodes with group information
nodes = [
    (0, 0),
    (1, 0),
    (2, 1),
    (3, 1),
    (4, 2),
    (5, 2)
]

for node_id, group in nodes:
    # Set node color based on its group
    color = group_color_mapping.get(group, 'gray')  # Default to gray if group not found
    g.add_node(node_id, group=group, color=color)

# Add edges
edges = [(0, 1), (2, 1), (3, 4), (3, 2), (4, 5), (1, 4), (1, 5)]
g.add_edges(edges)


"""
g.add_node(0, group=0)
g.add_node(1, group=0)
g.add_node(2, group=1)
g.add_node(3, group=1)
g.add_node(4, group=2)
g.add_node(5, group=2)
g.add_edges([(0, 1), (2,1), (3,4), (3,2), (4,5), (1,4), (1,5)])
"""

g.show_buttons(filter_=True)#['nodes', 'edges', 'physics'])

options = """
var options = {
  "configure":{
    "enabled": true
  },
  "nodes": {
    "borderWidth": 7,
    "borderWidthSelected": 2,
    "opacity": null,
    "size": 14
  },
  "edges": {
    "color": {
      "inherit": true
    },
    "selfReferenceSize": null,
    "selfReference": {
      "angle": 0.7853981633974483
    },
    "smooth": false
  },
  "layout": {
    "hierarchical": {
      "enabled": true
    }
  },
  "physics": {
    "hierarchicalRepulsion": {
      "centralGravity": 0,
      "avoidOverlap": null
    },
    "minVelocity": 0.75,
    "solver": "hierarchicalRepulsion"
  }
}
"""
#g.set_options(options)

g.write_html("graph_pyvis_test.html")
# Display the graph
#g.show("graph_pyvis_test.html")
```
